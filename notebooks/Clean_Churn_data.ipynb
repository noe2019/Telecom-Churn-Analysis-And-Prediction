```python
# Display numerical features
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
df[numeric_columns].head()
```

```python
# Display the number of missing values in each numerical column
df[numeric_columns].isnull().sum()
```

There is no need of handling missing values in the numerical columns, given that there is none

```python
# Dispay the columns with categorical values
categorical_columns = df.select_dtypes(include=['object']).columns
df[categorical_columns].head()
```

```python
# Fill missing categorical data with 'Unknown'
df[categorical_columns]= df[categorical_columns].fillna('Unknown')
df[categorical_columns].head()
```

```python
# Combine the numeric and filled categorical columns
df_combined = pd.concat([df[numeric_columns], df[categorical_columns]], axis=1)
df_combined.head()
```

```python
df_combined.isnull().sum()
```

```python
# Step 2: Correct Negative Values by using the absolute function
# Assuming negative values in 'Monthly_Charge' do not make sense
df_combined['Monthly_Charge'].unique()
```

```python
df_combined['Monthly_Charge'] = df_combined['Monthly_Charge'].abs()
```

```python
# Step 4: Convert Data Types
# Convert 'Total_Charges', 'Total_Refunds', 'Total_Extra_Data_Charges', 'Total_Long_Distance_Charges', and 'Total_Revenue' to numeric
df_combined[['Total_Charges', 'Total_Refunds', 'Total_Extra_Data_Charges', 'Total_Long_Distance_Charges', 'Total_Revenue']] = df_combined[['Total_Charges', 'Total_Refunds', 'Total_Extra_Data_Charges', 'Total_Long_Distance_Charges', 'Total_Revenue']].apply(pd.to_numeric, errors='coerce')

# Display cleaned structure
print("Cleaned DataFrame Structure:")
print(df_combined.info())

# Save cleaned data to a new CSV
cleaned_file_path = '../Data/Cleaned_Customer_Data.csv'
df_combined.to_csv(cleaned_file_path, index=False)
print(f"Cleaned data saved to {cleaned_file_path}")
```
